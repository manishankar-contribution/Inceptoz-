```python
#importing required packages
import pandas as pd
import datetime as datetime
```

```python
#reading as csv file using pandas
df=pd.read_csv('D:/Data Science/covid_test.csv')
```

```python
#printing out first 100 records
df.head(100)
```

```python
#converting the datetime formmating which we have in raw data like 03-11-2020 & 2/25/2020
df['Date']=pd.to_datetime(df.Date)
```

```python
#checking date has been changed as expected
df.head(100)
```

```python
df.Date.head(10)
```

```python
"""syntax to drop a non required cols.. axis =1 denotes Columns (which means in Y axis in terms in of graph view, axis =0 means
rows. you can assume as x axis in graph to remember easily)"""
df=df.drop(['Prep_Flow_Runtime','Latest_Date','Lat','Long'],axis=1)
```

```python
#sorting the data by date
df=df.sort_values(by=['Date'])
```

```python
#pivoting the dataframe as the column Case_Type has four values [Active Confirmed Deaths Recovered] need to tranform to columnar manner
import numpy as np
df2=pd.pivot_table(df,index=['Date','Country_Region','Province_State'],columns='Case_Type',values='Cases',aggfunc = np.sum)
#df2=df.pivot(columns=('Country_Region','Case_Type'),values='Cases')
```

```python
#Oops the pivoting has been done !!!
df2.head(100)
```

```python
#writing the output to csv file
df2.to_csv('D:/Data Science/covid_py2.csv')
```
